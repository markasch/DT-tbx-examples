```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the cobweb plotting function

```python
def cobweb(f,a,b,x0,N):
    x=np.linspace(a,b,N)
    y=f(x)
    plt.plot(x,x,'red')  # plot the line g(x)=x 
    plt.plot(x,y,'blue') # plot the curve f 
    for i in range(N):
        s=x0
        x0=f(x0)
        plt.plot([s,x0],[x0,x0],'black') # project point (x,f(x)) onto line (x,x) 
        plt.plot([x0,x0],[x0,f(x0)],'black') # project point (x0,x0) onto next iterate
```

Define the function we want to study

```python
def f1(x):
    return 2*x*(1-x)
```

Plot the cobweb from an inital point $x_0 = 0.1.$

```python
cobweb(f1,0,1,0.1,100)
```
