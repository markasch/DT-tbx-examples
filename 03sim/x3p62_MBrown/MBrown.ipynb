```python
import numpy as np
import matplotlib.pyplot as plt
# time interval and time-step
T = 1
N = 500
dt = T/N
t = np.linspace(0,1,N)
# initialize
dW = np.zeros(N)
W  = np.zeros(N)
# first step
dW[0] = np.sqrt(dt)*np.random.randn()
W[0]  = dW[0]
# loop over time
for j in range(N-1):
    dW[j+1] = np.sqrt(dt)*np.random.randn()
    W[j+1]  = W[j] + dW[j+1]
# plot the BM path
plt.plot(t,np.append([0], W[1:N]))
plt.xlabel('t'); plt.ylabel('W(t)')
```

Here is a vectorized version, where we replace the loop over time by a single vector operation, `cumsum`.

```python
# initialize
dW = np.zeros(N)
W  = np.zeros(N)
# all steps
dW = np.sqrt(dt)*np.random.randn(N)
W  = np.cumsum(dW)
# plot the BM path
plt.plot(t,np.append([0], W[1:N]))
plt.xlabel('t'); plt.ylabel('W(t)')
```
